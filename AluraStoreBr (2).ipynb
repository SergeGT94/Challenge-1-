# Task
Analyze the provided sales data from four different stores to identify the most and least sold products in each store. Combine the data from all stores, group by store and product to count sales, and then determine the most and least sold products for each store. Finally, visualize the results to clearly show the top and bottom performing products in terms of sales for each store. The data is available in four separate CSV files: "store1_sales.csv", "store2_sales.csv", "store3_sales.csv", and "store4_sales.csv".

## Combine data

### Subtask:
Concatenate the data from all four stores into a single DataFrame.


**Reasoning**:
Concatenate the data from all four stores into a single DataFrame.



```python
all_stores = pd.concat([loja, loja2, loja3, loja4])
```

## Group and count sales

### Subtask:
Group the combined data by store and product, then count the number of sales for each product in each store.


**Reasoning**:
Group the combined data by store and product, count the sales, reset the index, and rename the count column.



```python
sales_by_store_product = all_stores.groupby(['Local da compra', 'Produto']).size().reset_index(name='sales_count')
display(sales_by_store_product.head())
```

## Identify most and least sold

### Subtask:
For each store, determine the products with the highest and lowest sales counts.


**Reasoning**:
For each store, find the product with the highest and lowest sales counts and store them in separate dataframes.



```python
most_sold_products = sales_by_store_product.loc[sales_by_store_product.groupby('Local da compra')['sales_count'].idxmax()]
least_sold_products = sales_by_store_product.loc[sales_by_store_product.groupby('Local da compra')['sales_count'].idxmin()]
display(most_sold_products.head())
display(least_sold_products.head())
```

## Visualize results

### Subtask:
Create visualizations (e.g., bar charts) to display the most and least sold products for each store.


**Reasoning**:
Create bar plots to visualize the most and least sold products per store as requested.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
plt.bar(most_sold_products['Local da compra'], most_sold_products['sales_count'])
plt.title("Most Sold Products per Store")
plt.xlabel("Store")
plt.ylabel("Sales Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 6))
plt.bar(least_sold_products['Local da compra'], least_sold_products['sales_count'])
plt.title("Least Sold Products per Store")
plt.xlabel("Store")
plt.ylabel("Sales Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Análisis de Datos de Ventas de Tiendas

Este repositorio contiene el análisis de datos de ventas de cuatro tiendas diferentes. El objetivo principal es analizar el rendimiento de cada tienda en función de varios factores clave para identificar la tienda más adecuada para que el Sr. Juan venda sus productos.

## Datos

El conjunto de datos se compone de cuatro archivos CSV, uno por cada tienda:

*   `loja_1.csv`
*   `loja_2.csv`
*   `loja_3.csv`
*   `loja_4.csv`

Cada archivo CSV contiene las siguientes columnas:

*   **Produto**: Nombre del producto vendido.
*   **Categoria do Produto**: Categoría a la que pertenece el producto.
*   **Preço**: Precio del producto.
*   **Frete**: Costo de envío del producto.
*   **Data da Compra**: Fecha en que se realizó la compra.
*   **Vendedor**: Nombre del vendedor.
*   **Local da compra**: Ubicación de la tienda (estado).
*   **Avaliação da compra**: Calificación de la compra por parte del cliente.
*   **Tipo de pagamento**: Método de pago utilizado.
*   **Quantidade de parcelas**: Número de cuotas si el pago fue con tarjeta de crédito.
*   **lat**: Latitud de la ubicación de la compra.
*   **lon**: Longitud de la ubicación de la compra.

## Análisis Realizado

Se llevaron a cabo los siguientes análisis:

1.  **Análisis del faturamento**: Cálculo del ingreso total para cada tienda.
2.  **Ventas por Categoría**: Identificación de las categorías de productos más vendidas en todas las tiendas.
3.  **Média de Avaliação das Lojas**: Cálculo de la calificación promedio de los clientes para cada tienda.
4.  **Produtos Mais e Menos Vendidos**: Identificación de los productos más y menos vendidos en cada tienda.
5.  **Frete Médio por Loja**: Cálculo del costo de envío promedio para cada tienda.
